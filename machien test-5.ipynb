```python
# 5.Using Matplotlib or Seaborn, create a histogram to visualize the distribution of a
# numerical variable in a dataset.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filepath="C:/Users/afsha/Videos/Wine_Quality.csv"
df=pd.read_csv(filepath)
df
```
